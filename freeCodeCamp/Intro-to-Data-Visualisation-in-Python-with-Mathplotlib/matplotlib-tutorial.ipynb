#### Add another line

```python
x2 = np.arange(0,4.5,0.5)
print(x2)
```

```python
# Shorthand notation
# fmt = "{colour}{marker}{line}"

plt.plot(x, y, "b^--", label="2x")

plt.plot(x2, x2**2)

plt.title("Our First Graph", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
# Shorthand notation
# fmt = "{colour}{marker}{line}"

plt.plot(x, y, "b^--", label="2x")

plt.plot(x2, x2**2, "r", label="x^2")

plt.title("Our First Graph", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

```python
# Shorthand notation
# fmt = "{colour}{marker}{line}"

plt.plot(x, y, "b^--", label="2x")

plt.plot(x2[:6], x2[:6]**2, "r", label="x^2")
plt.plot(x2[5:], x2[5:]**2, "r--")

plt.title("Our First Graph", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

#### Resize a Graph

```python
# Shorthand notation
# fmt = "{colour}{marker}{line}"

plt.figure(figsize=(5,3), dpi=300)

plt.plot(x, y, "b^--", label="2x")

plt.plot(x2[:6], x2[:6]**2, "r", label="x^2")
plt.plot(x2[5:], x2[5:]**2, "r--")

plt.title("Our First Graph", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()

plt.show()
```

#### To Save a Graph

```python
# Shorthand notation
# fmt = "{colour}{marker}{line}"

plt.figure(figsize=(5,3), dpi=300)

plt.plot(x, y, "b^--", label="2x")

# Plot part of graph as a line
plt.plot(x2[:6], x2[:6]**2, "r", label="x^2")
#Plot remainder as a dot
plt.plot(x2[5:], x2[5:]**2, "r--")

#Add a title
plt.title("Our First Graph", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})

# X and Y labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# X and Y axes tick marks
plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

# Add a legend
plt.legend()

# Save figure
plt.savefig("mygraph.png", dpi=200)
plt.savefig("mygraph.svg", dpi=200)

# Show the graph plot
plt.show()
```

### Bar Chart

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

plt.bar(labels, values)

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

plt.bar(labels, values)

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

bars = plt.bar(labels, values)

bars[0].set_hatch("/")

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

bars = plt.bar(labels, values)

bars[0].set_hatch("/")
bars[1].set_hatch("o")

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

bars = plt.bar(labels, values)

bars[0].set_hatch("/")
bars[1].set_hatch("o")
bars[2].set_hatch("*")

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

bars = plt.bar(labels, values)

patterns = ["/", "o", "*"]
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

bars = plt.bar(labels, values)

patterns = ["/", "o", "*"]
for bar in bars:
    bar.set_hatch(patterns.pop())

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

bars = plt.bar(labels, values)

patterns = ["/", "o", "*"]
for i, bar in enumerate(bars):
    bar.set_hatch(patterns[i])

plt.figure(figsize=(6,4))

plt.show()
```
